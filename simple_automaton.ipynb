```python
def automaton_code(word: str):
    state = 'q0'
    for letter in word:
        print(state + '-' + letter + '->', end='')
        if state == 'q0':
            if letter == 'a':
                state = 'q1'
            elif letter == 'b':
                state = 'q3'
            else:
                raise ValueError('Invalid letter')
        elif state == 'q1':
            if letter == 'a':
                state = 'q0'
            elif letter == 'b':
                state = 'q2'
            else:
                raise ValueError('Invalid letter')
        elif state == 'q2':
            if letter == 'a':
                state = 'q3'
            elif letter == 'b':
                state = 'q1'
            else:
                raise ValueError('Invalid letter')
        elif state == 'q3':
            if letter == 'a':
                state = 'q2'
            elif letter == 'b':
                state = 'q0'
            else:
                raise ValueError('Invalid letter')
        else:
            raise ValueError('Invalid state')
    print(state)

    if state == 'q0':
        print("Word accepted")
    else:
        print("Word not accepted")
```

```python
tLetter = str
tState = str

Q: set[tState] = {'q0', 'q1', 'q2', 'q3'}
q0: tState = 'q0'
delta: dict[tuple[tState, tLetter], tState] = {
    ('q0', 'a'): 'q1',
    ('q0', 'b'): 'q3',
    ('q1', 'a'): 'q0',
    ('q1', 'b'): 'q2',
    ('q2', 'a'): 'q3',
    ('q2', 'b'): 'q1',
    ('q3', 'a'): 'q2',
    ('q3', 'b'): 'q0',
}
F: set[tState] = {'q0'}

def automaton_table(word: str):
    state = q0
    for letter in word:
        print(state + '-' + letter + '->', end='')
        state = delta[(state, letter)]

    print(state)
    if state in F:
        print("Word accepted")
    else:
        print("Word not accepted")
```
